## Perguntas de partida e hipóteses

1 - Marcas.

2 - Tipos (modelo) de carro.

3 - Relação marcas x tipo_combustivel. 

```python
df = df.rename(columns={'fuel-type': 'Combustivel'})
```

```python
marcas = df.groupby('make')['make'].count()
modelos = df.groupby('body-style')['body-style'].count()
df_gb = df.groupby(['make','Combustivel']).size().unstack()
```

```python
marcas
```

```python
modelos
```

```python

df_gb.plot(kind = 'bar', figsize=(10,5), rot=45)

plt.title('Quantidade de Marcas', fontsize=14)
plt.xlabel('Marcas', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)
plt.grid(True)

plt.show()
```

## Insights

```python
fig, ax = plt.subplots(figsize=(10,5))
modelos.plot(kind = 'bar', color='red')
plt.setp(ax.get_xticklabels(), rotation=45)
plt.title('Quantidade de Modelo', fontsize=14)
plt.xlabel('Modelos', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)
plt.grid(True)

plt.show()
```
